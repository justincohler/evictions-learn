```python

# define models to run
models_to_run=['RF','LR', 'NB', 'GB', 'DT', 'KNN'] #'SVM'

# select features to use
features  =  ['geo_id', 'population', 'evictions', 'eviction_filings', 'eviction_filing_rate', 'renter_occupied_households']

classifiers, params = pipeline.get_classifiers()
df = pipeline.fill_missing(df)

X = df[features]
y = df['top20_rate']

baselines_to_run=['RAND']

results_df = pipeline.classify(models_to_run, classifiers, params, X,y, baselines_to_run)

results_df.sort_values(by=['auc-roc'], ascending=False)
```
